```python
import pandas as pd
import os
df = pd.read_excel("./Population demographics (1).xlsx")
df
```

```python
print(df.duplicated())
```

```python
df.dropna(inplace = True)
df
```

```python
df = df.drop(df[df['Geography'] == 'Canada'].index)
df
```

```python
newdf = df.melt(id_vars = ['Geography'],
        var_name = 'Date', 
         value_name = 'Population' )
newdf
```

```python
df1=newdf.replace({'Geography':{'Alberta':'ALBERTA','British Columbia':'BRITISH_COLUMBIA','Manitoba':'MANITOBA','New Brunswick':'NEW_BRUNSWICK','Newfoundland and Labrador':'NEWFOUNDLAND_AND_LABRADOR', 'Nova Scotia':'NOVA_SCOTIA','Ontario': 'ONTARIO','Prince Edward Island': 'PRINCE_EDWARD_ISLAND','Quebec':'QUEBEC','Saskatchewan':'SASKATCHEWAN'}})
df1
```

```python
df1.rename(columns={'Geography':'Location'},inplace = True)
df1
```

```python
columns_titles = ["Date","Location","Population"]
df1=df1.reindex(columns=columns_titles)
df1
```

```python
df1['Date'] = df1['Date'].apply(lambda x: pd.to_datetime(str(x), format='%Y'))
```
