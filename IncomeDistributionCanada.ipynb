```python
df.isnull().values.any()
df.head
```

dropping 'Null' values

```python
df.dropna(inplace = True)
df
```

Reshaping the DataFrame by turning rows into columns

```python
df1 = df.melt(id_vars = ['year', 'avg_employment_income', 'median_employment_income', 'population_with_income'],
        var_name = 'IncomeDistributionInCad', 
         value_name = 'PercentIncomeDistribution' )
df1
```

Renaiming the columns for clearer understanding

```python
df1.rename(columns={'year':'Year','avg_employment_income':'AVGemploymentIncome','median_employment_income':'MedianEmploymentIncome','population_with_income':'PopulationWithIncome'},inplace = True)
df1
```

Creating Bar Chart to visualise specific components of the DataFrame

```python
import matplotlib.pyplot as plt
```

```python
x = df1['Year']
y = df1['AVGemploymentIncome'] 
plt.xlabel('AVGemploymentIncome')
plt.ylabel('Year')
plt.barh(x,y)
```
